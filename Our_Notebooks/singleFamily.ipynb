```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

```

```python
# Create a reference to the CSV and import it into a Pandas DataFrame
csv_path = Path("Resources/State_zhvi_uc_sfr_tier_0.33_0.67_sm_sa_month_.csv")
price_df = pd.read_csv(csv_path)
price_df.head()
```

```python
# Drop 'RegionID' and 'SizeRank' columns
home_price_df = price_df.drop(['RegionID', 'SizeRank', 'RegionType', 'StateName'], axis=1)

home_price_df.head()
#print(rate_df.columns)
```

```python
# # Drop columns from 2000-01-31 to 2014-12-31
new_df = home_price_df.drop(home_price_df.columns[1:181], axis=1)

new_df.head()
```

```python
# Plotting the data
plt.figure(figsize=(12, 6))

# Extract the years from the column names
years = new_df.columns[1:]
x_axis = np.arange(len(years))


# Plotting the data
# Select one data point for each year
selected_years = ['2015-12-31', '2016-12-31', '2017-12-31', '2018-12-31', '2019-12-31', '2020-12-31', '2021-12-31', '2022-12-31']
selected_data = new_df[['RegionName'] + selected_years]

# Plotting the scatter diagram for California
plt.figure(figsize=(12, 6))
california_prices = selected_data[selected_data['RegionName'] == 'California'].values.flatten()[1:].astype(float)
plt.scatter(selected_years, california_prices, label='California')

plt.xlabel('Years')
plt.ylabel('Home Price')
plt.title('Home Price in California')
plt.legend()

plt.show()
```
