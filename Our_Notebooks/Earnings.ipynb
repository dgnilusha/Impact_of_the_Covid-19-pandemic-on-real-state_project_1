```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from scipy.stats import linregress
```

```python
e_path = ("data/FRED_earnings.csv")
earning_df = pd.read_csv(e_path)
earned_df = earning_df.rename(columns = {"CAEREA": "CA Real Estate Earnings(Thousands of Dollars)"})
earned_df.head()
```

```python
#Plot Data
#Set axes
x_values = [2017.833, 2018.083, 2018.333, 2018.583, 2018.833,
            2019.083, 2019.333, 2019.583, 2019.833,
           2020.083, 2020.333, 2020.583, 2020.833,
           2021.083, 2021.333, 2021.583, 2021.833,
           2022.083, 2022.333, 2022.583, 2022.833]
y_values = earned_df["CA Real Estate Earnings(Thousands of Dollars)"]

#Retrieve line regression information
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = []
for x in x_values:
    regress = slope * x + intercept
    regress_values.append(regress)
    
line_eq = "y= " + str(round(slope,2)) + "x +" + str(round(intercept,2))
plt.plot(x_values, regress_values, "r-")
plt.annotate(line_eq, (2018, 65000000), color = "red")
ax=plt.gca()
ax.ticklabel_format(style='plain')

plt.scatter(x_values, y_values, marker='o', label = "California" )
plt.xlabel("DATE")
plt.ylabel("CA Real Estate Earnings(Thousands of Dollars)")
plt.title("CA Rental and Leasing Earnings in California")
plt.legend()
plt.show()
    
print("Regression Line Equation: y = {:.2f}x + {:.2f}".format(slope, intercept))
print("Correlation Coefficient:", rvalue)
```
