```python
# Create a DataFrame from the provided dataset
data = {"Year":[2015, 2016,2017, 2018, 2019, 2020, 2021, 2022 ], 
        "Population": [38918045, 39167117, 39358497, 39461588, 39512223, 39501653, 39142991, 39029342]}

newdf = pd.DataFrame(data)
newdf
```

```python
# # Plotting the line plot
plt.plot(newdf['Year'], newdf['Population'], marker='o')
plt.xlabel('Year')
plt.ylabel('Resident Population')
plt.title('California Resident Population Trend')
plt.grid(True)
plt.show()
```

```python

Year = newdf['Year']
Resident_Population = newdf['Population']

# Perform linear regression for California Resident Population Trend
slope, intercept, r_value, p_value, std_err = linregress(Year, Resident_Population)
line_eq = "y = {:.2f}x + {:.2f}".format(slope, intercept)

# Calculate regression line for California Resident Population Trend
regression_line = slope * Year + intercept

# Plotting the line chart with regression line for California Resident Population Trend
plt.figure(figsize=(12, 6))
plt.plot(Year, Resident_Population, marker='o', linestyle='-', label='Resident Population')
plt.plot(Year, regression_line, linestyle='--', label='Regression Line')
plt.annotate(line_eq, (2020, 38918045), fontsize=15, color="red")

plt.xlabel('Year')
plt.ylabel('Resident Population')
plt.title('California Resident Population Trend')
plt.grid(True)
plt.legend()
plt.show()

print("Regression Line Equation (California Resident Population Trend): y = {:.2f}x + {:.2f}".format(slope, intercept))
print("Correlation Coefficient (California Resident Population Trend):", r_value)
```
