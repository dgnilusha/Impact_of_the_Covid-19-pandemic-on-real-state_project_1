```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Create a reference to the CSV and import it into a Pandas DataFrame
csv_path = Path("Resources/List_&_Sale_price.monthly.csv")
#Desktop/Project_1/List and sale price/Resources/List_&_Sale_price.monthly.csv
list_and_sale_df = pd.read_csv(csv_path)
list_and_sale_df.head()
```

```python
# clean data
# Drop 'RegionID' and 'SizeRank' columns
Clean_list_and_sale_df = list_and_sale_df.drop(['SizeRank', 'RegionID', 'RegionType', ], axis=1)

Clean_list_and_sale_df.head()
#print(Clean_list_and_sale_df)
#print(rate_df.columns)

```

```python
# # Drop columns from 2000-01-31 to 2014-12-31
Clean2_list_and_sale_df = Clean_list_and_sale_df.drop(Clean_list_and_sale_df.columns[48:64], axis=1)

Clean2_list_and_sale_df.head()
```

```python


# Extract data for San Jose, Los Angeles, and San Francisco
san_jose_data = Clean2_list_and_sale_df[Clean_list_and_sale_df['RegionName'] == 'San Jose, CA'].iloc[:, 2:]
los_angeles_data = Clean2_list_and_sale_df[Clean_list_and_sale_df['RegionName'] == 'Los Angeles, CA'].iloc[:, 2:]
san_francisco_data = Clean2_list_and_sale_df[Clean_list_and_sale_df['RegionName'] == 'San Francisco, CA'].iloc[:, 2:]
sacramento_data = Clean2_list_and_sale_df[Clean_list_and_sale_df['RegionName'] == 'Sacramento, CA'].iloc[:, 2:]

# Convert the date columns to datetime format
san_jose_data.columns = pd.to_datetime(san_jose_data.columns)
los_angeles_data.columns = pd.to_datetime(los_angeles_data.columns)
san_francisco_data.columns = pd.to_datetime(san_francisco_data.columns)
sacramento_data.columns = pd.to_datetime(sacramento_data.columns)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(san_jose_data.columns, san_jose_data.values.flatten(), label='San Jose')

#plt.plot(date, regression_line, linestyle='--', label='Regression Line')

plt.plot(los_angeles_data.columns, los_angeles_data.values.flatten(), label='Los Angeles')
plt.plot(san_francisco_data.columns, san_francisco_data.values.flatten(), label='San Francisco')
plt.plot(sacramento_data.columns, sacramento_data.values.flatten(), label='Sacramento')

plt.grid(True)

plt.xlabel('Date')
plt.ylabel('Sale Prices')
plt.title('Sale Prices in San Jose, Los Angeles, San Francisco, sacramento')
plt.legend()
plt.show()

```

```python
#from scipy.stats import ttest_ind
import scipy.stats as stats
# t-test between San Jose and San Francisco

stats.ttest_ind(san_jose_data.values.flatten(), san_francisco_data.values.flatten(), equal_var=False)

# # Print the t-statistic and p-value
#print("T-Statistic:", t_stat)
#print("P-Value:", p_value)
```
